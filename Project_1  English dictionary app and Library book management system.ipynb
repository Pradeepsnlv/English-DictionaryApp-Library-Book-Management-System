# 1B LIBRARY BOOK MANAGEMENT SYSTEM

```python
# Libraries

from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
```

```python
root=Tk()
root.title('Library Book Management System')
```

```python
# Methods

def Database():
    global conn,cursor
    conn=sqlite3.connect('books.db')
    cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS Booklibrary(Booktitle TEXT,Author TEXT,Year INTEGER,ISBN TEXT)')
    

def Create():
    if Booktitle.get()=='' or Author.get()=='' or Year.get()=='' or ISBN.get()=='':
        txt_result.config(text='Please enter all the fields',fg='red')
        
    else:
        Database()
        cursor.execute('INSERT INTO Booklibrary(Booktitle,Author,Year,ISBN) VALUES(?,?,?,?)',(str(Booktitle.get()),str(Author.get()),str(Year.get()),str(ISBN.get())))
        conn.commit()
        Booktitle.set("")
        Author.set("")
        Year.set("")
        #ISBN.set("")
        cursor.close()
        conn.close()
        txt_result.config(text='Record Inserted!!!',fg='green')
        
def Read():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute('SELECT * FROM Booklibrary ORDER BY Author ASC')
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    cursor.close()
    txt_result.config(text='Successfully fetched data',fg='black')
    
def Exit():
    result=tkMessageBox.askquestion('Do you want to quit?(y/n)',icon='warning')
    if result=='yes':
        root.destroy()
        exit

    
    
def Search():
        if Booktitle.get()=='' :
            txt_result.config(text='Booktitle field is mandatory!!!!',fg='red')
            
        else:            
            tree.delete(*tree.get_children())
            conn=sqlite3.connect('books.db')
            cursor=conn.execute("SELECT * FROM Booklibrary WHERE Booktitle LIKE ?",('%'+str(Booktitle.get())+'%',))
            fetch=cursor.fetchall()
            for data in fetch:
                tree.insert('','end',values=(data))
            cursor.close()
            conn.close()
            txt_result.config(text='Search successful',fg='green')  
        
     
    
def Delete():
        if Booktitle.get()=='' :
            txt_result.config(text='Booktitle field is mandatory!!!!',fg='red')
            
        else:            
            tree.delete(*tree.get_children())
            conn=sqlite3.connect('books.db')
            cursor=conn.execute("DELETE FROM Booklibrary where Booktitle LIKE ?",('%'+str(Booktitle.get())+'%',))
            fetch=cursor.fetchall()
            for data in fetch:
                tree.insert('','end',values=(data))
            conn.commit()
            cursor.close()
            conn.close()
            txt_result.config(text='Deleted successfully',fg='green')     
    
```

```python
# Variables

Booktitle = StringVar()
Author = StringVar()
Year = StringVar()
ISBN = StringVar()
```

```python
# Frames 

Top = Frame(root, width=900, height=50, bd=8, relief='raise',bg='blue')
Top.pack(side=TOP)

Left= Frame(root, width=300, height=500, bd=8, relief='raise',bg='red')
Left.pack(side=LEFT)

Right= Frame(root, width=600, height=500, bd=8, relief='raise',bg='yellow')
Right.pack(side=RIGHT)

Forms = Frame(Left, width=300, height=450)
Forms.pack(side=TOP)
```

```python
# Buttons

Buttons= Frame(Left, width=300, height=100, bd=8, relief='raise')
Buttons.pack(side=BOTTOM)
```

```python
# Labels

txt_title = Label(Top,width=900,text='Library Book Management System',font=('arial',16))
txt_title.pack()

txt_booktitle= Label(Forms,text='Booktitle',font=('arial',16),bd=15)
txt_booktitle.grid(row=0,stick='e')

txt_author= Label(Forms,text='Author',font=('arial',16),bd=15)
txt_author.grid(row=1,stick='e')

txt_year= Label(Forms,text='Year',font=('arial',16),bd=15)
txt_year.grid(row=2,stick='e')

txt_ISBN= Label(Forms,text='ISBN',font=('arial',16),bd=15)
txt_ISBN.grid(row=3,stick='e')

txt_result = Label(Buttons)
txt_result.pack(side=TOP)
```

```python
# Entry Widget

booktitle = Entry(Forms, textvariable=Booktitle,width=30)
booktitle.grid(row=0,column=1)

author = Entry(Forms, textvariable=Author,width=30)
author.grid(row=1,column=1)

year = Entry(Forms, textvariable=Year,width=30)
year.grid(row=2,column=1)

ISBN = Entry(Forms, textvariable=ISBN,width=30)
ISBN.grid(row=3,column=1)
```

```python
# Button widget

btn_display = Button(Buttons,width=10,text='Display',command=Read)
btn_display.pack(side=LEFT)

btn_search = Button(Buttons,width=10,text='Search',command=Search)
btn_search.pack(side=LEFT)

btn_add = Button(Buttons,width=10,text='Add',command=Create)
btn_add.pack(side=LEFT)

btn_issue = Button(Buttons,width=10,text='Issue')
btn_issue.pack(side=LEFT)

btn_delete = Button(Buttons,width=10,text='Delete',command=Delete)
btn_delete.pack(side=LEFT)

btn_exit = Button(Buttons,width=10,text='Exit',command=Exit)
btn_exit.pack(side=LEFT)
```

```python
# List widget

scrollbary = Scrollbar(Right, orient=VERTICAL)
scrollbarx = Scrollbar(Right, orient=HORIZONTAL)

tree = ttk.Treeview(Right, columns=('Booktitle','Author','Year','ISBN',))

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT, fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM, fill=X)

tree.heading('Booktitle',text='Booktitle',anchor=W)
tree.heading('Author',text='Author',anchor=W)
tree.heading('Year',text='Year',anchor=W)
tree.heading('ISBN',text='ISBN',anchor=W)

tree.column('#0',stretch=NO,minwidth=0,width=0)
tree.column('#1',stretch=NO,minwidth=0,width=80)
tree.column('#2',stretch=NO,minwidth=0,width=120)
tree.column('#3',stretch=NO,minwidth=0,width=80)
tree.pack()
```

# 1A ENGLISH DICTIONARY APP

```python
import json
```

```python
d1 = {1:2}
f = open('words.txt','w')
json.dump(d1,f)
f.close()
```

```python
# Methods

def Add():
    word = input('Enter a New Word :')
    meaning= input('Enter the meaning of the word which you have entered :')
    f = open('words.txt','r')
    n = json.load(f)
    f.close()
    n[word] = meaning
    f = open('words.txt','w')
    json.dump(n,f)
    f.close()
    print('Word Dumped Successfully!!!')
    
    
    
def Finding():
    f=open('words.txt','r')
    n=json.load(f)
    f.close()
    find=input("Enter the word to find its meaning : ")
    for k,v in n.items():
        if k==find:
            print("Meaning of entered word is :",v)
            break
                       
    else:
        print('No such word exist in dictionary') 
            
            
def Update():
    f = open('words.txt','r')
    n = json.load(f)
    f.close
    word1 = input('Enter the word whose meaning is to be updated :')
    mean = input('Enter the new meaning of the word :')
    
    for k,v in n.items():
        if k == word1:
            n[k] = mean
            f = open('words.txt','w')
            json.dump(n,f)
            f.close()
            print('Successfully updated the meaning!!!')
            break
            
    else:
        print('No such word is available in dictionary...')
        
        

while True:
    
    print('\n')
    print('Main menu \n\n  1. Add a Word \n 2.Find the meaning \n 3. Update a word \n 4.Exit \n')
    try:
        action = int(input('Please select from the given mainmenu(1,2,3,4) :'))
    except ValueError:
        print('!!!!Please enter integer values only!!!!')
        
        
    else:
        if action == 1:
            Add()
            
        elif action == 2:
            Finding()
            
        elif action == 3:
            Update()
            
        elif action == 4:
            e = input('Are you sure do you want to exit ?(y/n) :')
            if e == 'yes' or e == 'Yes':
                print('Thank you Have a nice day!!!')
                break
                
        else:
            print('Please choose the correct option from the mainmenu!!!')
            
```
